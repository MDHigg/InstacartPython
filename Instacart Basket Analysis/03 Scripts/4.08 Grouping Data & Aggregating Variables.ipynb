##### d) Aggregating Data with transform()

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(max)
```

#### e) Loyal customer column usind .loc()

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.head()
```

```python
#counting number of customer types
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

### 07 Create an order frequency flag that marks the regularity of a user’s ordering behavior

```python
# create a column with the median orders
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform('median')
```

```python
ords_prods_merge['order_frequency']
```

```python
# create a column with the median orders

ords_prods_merge.loc[ords_prods_merge['order_frequency'] >= 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] < 20) & (ords_prods_merge['order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# check that these additions have been made to the dataframe
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# for customers with only 1 order there is no media value for days since last order
# these customers need to be labelled as 'non-frequent customers'
# use fillna() to replace these and then value_counts to check for NaN and non-frequent customer values

ords_prods_merge['order_frequency_flag'].fillna('Non-frequent customer', inplace = True)
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
